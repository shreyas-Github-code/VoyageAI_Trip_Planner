```python
# Simple Todo List Application
class TodoList:
    def __init__(self):
        self.todos = []
    
    def add_todo(self, task):
        """Add a new task to the todo list"""
        todo = {
            "task": task,
            "completed": False
        }
        self.todos.append(todo)
        print(f"✓ Added: {task}")
    
    def mark_completed(self, index):
        """Mark a task as completed"""
        try:
            if 0 <= index < len(self.todos):
                self.todos[index]["completed"] = True
                print(f"✓ Completed: {self.todos[index]['task']}")
            else:
                print("Error: Invalid task number")
        except:
            print("Error: Could not mark task as completed")
    
    def remove_todo(self, index):
        """Remove a task from the list"""
        try:
            if 0 <= index < len(self.todos):
                removed = self.todos.pop(index)
                print(f"✓ Removed: {removed['task']}")
            else:
                print("Error: Invalid task number")
        except:
            print("Error: Could not remove task")
    
    def display_todos(self):
        """Display all tasks"""
        if not self.todos:
            print("\nNo tasks yet!\n")
            return
        
        print("\n--- Your Todo List ---")
        for i, todo in enumerate(self.todos):
            status = "✓" if todo["completed"] else "○"
            print(f"{i}: {status} {todo['task']}")
        print()

# Create todo list
my_list = TodoList()

# Add some tasks
my_list.add_todo("Learn Python basics")
my_list.add_todo("Practice loops and functions")
my_list.add_todo("Build a small project")
my_list.add_todo("Learn about libraries")

# Display todos
my_list.display_todos()

# Mark first task as completed
my_list.mark_completed(0)

# Add more tasks
my_list.add_todo("Master error handling")

# Display updated todos
my_list.display_todos()

# Remove a task
my_list.remove_todo(1)

# Final display
print("Final todo list:")
my_list.display_todos()

print("✓ Mini project completed! You've learned the fundamentals of Python!")
```

```python
# Example 1: Division by zero error
print("Example 1: Handling division by zero")
try:
    result = 10 / 0
except ZeroDivisionError:
    print("  Error: Cannot divide by zero!")
    result = 0
print(f"  Result: {result}\n")

# Example 2: Index out of range error
print("Example 2: Handling index out of range")
my_list = [1, 2, 3]
try:
    print(f"  Item at index 5: {my_list[5]}")
except IndexError:
    print("  Error: Index out of range!")
print()

# Example 3: Type error
print("Example 3: Handling type error")
try:
    result = "hello" + 5
except TypeError:
    print("  Error: Cannot add string and integer!")
print()

# Example 4: Converting user input with error handling
print("Example 4: Converting user input with try-except")
try:
    number = int("123")  # This works
    print(f"  Successfully converted: {number}")
except ValueError:
    print("  Error: Could not convert to integer")

try:
    number = int("hello")  # This will fail
except ValueError:
    print("  Error: 'hello' cannot be converted to integer")
print()

# Example 5: Using else and finally
print("Example 5: Using else and finally blocks")
try:
    result = 10 / 2
    print(f"  10 / 2 = {result}")
except ZeroDivisionError:
    print("  Error: Cannot divide by zero")
else:
    print("  Code after try executed successfully (no error)")
finally:
    print("  This always executes, whether there's an error or not")
print()

# Example 6: Catching multiple exceptions
print("Example 6: Catching multiple exception types")
def process_data(data):
    try:
        if not isinstance(data, (int, float)):
            raise TypeError("Data must be a number")
        if data < 0:
            raise ValueError("Data must be positive")
        return data * 2
    except TypeError as e:
        print(f"  Type Error: {e}")
    except ValueError as e:
        print(f"  Value Error: {e}")
    except Exception as e:
        print(f"  Unexpected Error: {e}")

process_data(5)          # Works fine
process_data(-5)         # Value error
process_data("hello")    # Type error

print("\n✓ Error handling completed successfully!")
```

## Section 9: File Input and Output

Learn how to read from and write to files using Python's built-in file handling capabilities.

```python
# Simple function
def greet():
    print("Hello! Welcome to Python!")

greet()

# Function with parameters
def greet_person(name, age):
    print(f"Hello {name}! You are {age} years old.")

greet_person("Alice", 28)

# Function with default parameters
def introduce(name, city="Unknown"):
    print(f"{name} is from {city}")

introduce("Bob")  # Uses default
introduce("Charlie", "Boston")  # Uses provided value

# Function with return value
def add(a, b):
    return a + b

result = add(5, 3)
print(f"\n5 + 3 = {result}")

# Function returning multiple values
def divide_and_remainder(a, b):
    quotient = a // b
    remainder = a % b
    return quotient, remainder

q, r = divide_and_remainder(17, 5)
print(f"17 ÷ 5 = {q} remainder {r}")

# Function with variable number of arguments
def sum_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(f"\nSum of 1,2,3: {sum_numbers(1, 2, 3)}")
print(f"Sum of 1,2,3,4,5: {sum_numbers(1, 2, 3, 4, 5)}")

# Function with keyword arguments
def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print("\nDisplay info:")
display_info(name="David", age=35, city="NYC", job="Engineer")

# Docstrings (documentation for functions)
def calculate_area(length, width):
    """
    Calculate the area of a rectangle.
    
    Args:
        length: The length of the rectangle
        width: The width of the rectangle
    
    Returns:
        The area of the rectangle
    """
    return length * width

print(f"\nArea of rectangle (5x4): {calculate_area(5, 4)}")
print(f"Function docstring: {calculate_area.__doc__}")
```

## Section 8: Functions and Parameters

Functions are reusable blocks of code that perform specific tasks. They help organize and simplify your code.

```python
# For loop - iterate through a list
print("For loop - printing items:")
colors = ["red", "green", "blue", "yellow"]
for color in colors:
    print(f"  - {color}")

# For loop with range
print("\nFor loop with range (0-4):")
for i in range(5):
    print(f"  i = {i}")

# For loop with range (start, stop, step)
print("\nFor loop with range (1, 10, 2):")
for i in range(1, 10, 2):
    print(f"  i = {i}")

# For loop with index
print("\nFor loop with index:")
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits):
    print(f"  Index {index}: {fruit}")

# While loop
print("\nWhile loop - counting down:")
count = 5
while count > 0:
    print(f"  Count: {count}")
    count -= 1
print("  Blastoff!")

# While loop with break
print("\nWhile loop with break:")
i = 0
while True:
    print(f"  i = {i}")
    if i == 3:
        print("  Breaking out of loop!")
        break
    i += 1

# For loop with continue
print("\nFor loop with continue (skip even numbers):")
for i in range(1, 8):
    if i % 2 == 0:
        continue  # Skip even numbers
    print(f"  {i} is odd")

# Nested loops
print("\nNested loops - multiplication table:")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} × {j} = {i*j}", end="  ")
    print()  # New line
```

## Section 7: Loops (for and while)

Loops allow you to repeat code multiple times. Use 'for' loops for known iterations and 'while' loops for conditional repetition.

```python
# Simple if-else
age = 20
if age >= 18:
    print(f"Age {age} - You are an adult")
else:
    print(f"Age {age} - You are a minor")

# if-elif-else
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"\nScore {score} = Grade {grade}")

# Comparison operators
print("\nComparison operators:")
x = 10
y = 5
print(f"10 > 5: {x > y}")      # Greater than
print(f"10 < 5: {x < y}")      # Less than
print(f"10 == 5: {x == y}")    # Equal to
print(f"10 != 5: {x != y}")    # Not equal to
print(f"10 >= 5: {x >= y}")    # Greater than or equal
print(f"10 <= 5: {x <= y}")    # Less than or equal

# Logical operators
print("\nLogical operators (and, or, not):")
is_raining = True
is_cold = False

if is_raining and is_cold:
    print("It's raining AND cold - stay home!")
elif is_raining or is_cold:
    print("It's either raining OR cold")
else:
    print("Nice weather!")

if not is_raining:
    print("It's not raining anymore")

# Checking multiple conditions
name = "Charlie"
age = 25
city = "Boston"

if name == "Charlie" and age > 18 and city in ["Boston", "NYC"]:
    print(f"{name} from {city} is an adult and from a major city!")
```

## Section 6: Conditional Statements (if, elif, else)

Use conditionals to make decisions in your code based on different conditions.

```python
# Creating a dictionary
person = {
    "name": "Alice",
    "age": 28,
    "city": "New York",
    "is_student": False,
    "grades": [90, 85, 88]
}

print(f"Person dictionary: {person}")
print(f"Dictionary keys: {person.keys()}")
print(f"Dictionary values: {person.values()}")

# Accessing values
print(f"\nName: {person['name']}")
print(f"Age: {person['age']}")

# Adding new key-value pairs
person["phone"] = "555-1234"
print(f"\nAfter adding phone: {person}")

# Modifying values
person["age"] = 29
print(f"After updating age: {person['age']}")

# Removing key-value pairs
del person["is_student"]
print(f"\nAfter deleting 'is_student': {person}")

# Checking if key exists
if "name" in person:
    print(f"'name' exists in person")

# Iterating through dictionary
print("\nIterating through items:")
for key, value in person.items():
    print(f"  {key}: {value}")

# Get method (safer way to access with default value)
country = person.get("country", "Not specified")
print(f"\nCountry: {country}")  # Returns default if key doesn't exist
```

## Section 4: Lists and Basic Operations

Lists are ordered collections that can hold multiple items. They are mutable (can be changed).

```python
# String creation and basic operations
message = "Hello World"
print(f"Original: {message}")
print(f"Length: {len(message)}")

# String indexing (first character is at index 0)
print(f"\nFirst character: {message[0]}")
print(f"Last character: {message[-1]}")
print(f"Characters 0-4: {message[0:5]}")

# String methods
print(f"\nUppercase: {message.upper()}")
print(f"Lowercase: {message.lower()}")
print(f"Capitalize: {message.capitalize()}")

# String concatenation and formatting
first_name = "John"
last_name = "Doe"
age = 30

# Method 1: Concatenation
full_info = first_name + " " + last_name + " is " + str(age)
print(f"\nMethod 1 (Concatenation): {full_info}")

# Method 2: f-strings (recommended for Python 3.6+)
full_info = f"{first_name} {last_name} is {age}"
print(f"Method 2 (f-string): {full_info}")

# String methods
text = "python is awesome, python is powerful"
print(f"\nReplace 'python' with 'Python': {text.replace('python', 'Python')}")
print(f"Split by comma: {text.split(',')}")
```
